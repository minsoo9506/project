```python
import nltk
from collections import Counter
from konlpy.tag import Kkma
kkma=Kkma()
from konlpy.utils import pprint
```

```python
comment_file = open("10f_metoo_refined.csv", 'r')
comment_lines= comment_file.readlines()
comment_file.close()
```

### comment_lines를 twitter를 이용하여 품사태깅 -> doc_vec에 결과 저장

```python
from konlpy.tag import Twitter
tw=Twitter()
doc_vec = []
for i in range(0, len(comment_lines)):
    tokens = comment_lines[i].split(" ")
    word_vec = []
    for word in tokens:
        pos_join = ""
        for word_pos in tw.pos(word, norm=True, stem=True):
            if word_pos[1] == "Adjective" or word_pos[1]=="Noun":# or word_pos[1] == "Verb" 
                    pos_join += word_pos[0]
        if pos_join is not "":
            word_vec.append(pos_join)
    if word_vec!=[]:
        doc_vec.append(word_vec)
```

```python
words=[]
for doc in doc_vec:
    for w in doc:
        words.append(w)
```

```python
words
```

```python
count=Counter(words)
top_list=count.most_common(50)
```

```python
top_list
```

### top_list에 있는 단어들을 꼬꼬마로 형태소분석

```python
kkma_list=[]
for word,count in top_list:
    if word!="\n":
        for w,p in kkma.pos(word):
            kkma_list.append((w+'/'+p,count))
#             print(kkma.pos(w))
```

### kosac의 polarity를 이용해 긍/부정 판단

```python
import pandas as pd
file_polarity=pd.read_csv('polarity.csv')
```

```python
neg=0
pos=0
neut=0
comp=0
for index, row in file_polarity.iterrows():
    for wp,c in kkma_list:
        if str(file_polarity['ngram'].iloc[index]) == wp:
            if file_polarity['max.value'].iloc[index]=='POS':
                pos+=1*int(c)
                print(wp+','+str(c)+','+'POS')
            elif file_polarity['max.value'].iloc[index]=='NEG':
                neg-=1*int(c)
                print(wp+','+str(c)+','+'NEG')
            elif file_polarity['max.value'].iloc[index]=='NEUT':
                neut+=1*int(c)
                print(wp+','+str(c)+','+'NEUT')
            else:
                comp+=1*int(c)
                print(wp+','+str(c)+','+'COMP')
```

```python
print('neg:',neg,'pos:',pos,'neut:',neut,'comp:',comp)
```
