```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
kim = pd.read_csv('kim.csv')
```

```python
kim['pos/pos+neg'] = kim['pos'] / (kim['pos'] + kim['neg']) 
```

```python
x = np.arange(6)
neg = kim['neg']
pos = kim['pos']
neut = kim['neu']
comp = kim['com']
feel = kim['pos/pos+neg']
```
